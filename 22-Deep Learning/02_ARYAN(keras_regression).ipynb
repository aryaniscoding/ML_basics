#MODEL PREPARATION : NEXT VIDEO


```
X = df.drop('price',axis=1).values
y = df['price'].values
```

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```
# SCALING
```

```
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```
X_train = scaler.fit_transform(X_train)
```

```
X_test = scaler.transform(X_test)
```

```
from tensorflow.keras.models import Sequential
```

```
from tensorflow.keras.layers import Dense
```

```
X_train.shape
```

```
model = Sequential()
# add 19 layers in the dl model
```

```
model.add(Dense(19,activation='relu'))
```

```
model.add(Dense(19,activation='relu'))
```

```
model.add(Dense(19,activation='relu'))
```

```
model.add(Dense(19,activation='relu'))
```

```
model.add(Dense(1))
```

```
model.compile(optimizer='adam',loss='mse')
```

```
model.fit(x=X_train , y=y_train , validation_data=(X_test,y_test),
          batch_size = 128,epochs=400)
```

#Evaluating the model : Next Video   

```
loses = pd.DataFrame(model.history.history)
```

```
loses.plot()
```

HERE we see the training loss and the validation loss go down and continue down together hence there is no overfitting occuring here

hence no need of early stopping here


```
from sklearn.metrics import mean_absolute_error,mean_squared_error,explained_variance_score
```

```
predictions = model.predict(X_test)
```

```
mean_squared_error(y_test,predictions)
```

```
np.sqrt(mean_squared_error(y_test,predictions))
```

```
mean_absolute_error(y_test,predictions)
```

```
# this model doesnt predict correctly , it is off by approx 20%
# hence we perform the explained_variance_score to check the mistakes
```

```
explained_variance_score(y_test,predictions)
```

Best value we could have got here is 1.0 , so we are training our model again till it reaches a stage of overfitting


```
plt.figure(figsize=(12,6))
plt.scatter(y_test,predictions)
plt.plot(y_test,y_test,'r')
```
