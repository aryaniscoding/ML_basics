```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
#linear Model plot
sns.lmplot(x='total_bill',y='tip',data=tips,hue = 'sex',markers=['o','v'],scatter_kws={'s':50})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='smoker',hue='sex',row='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',row='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='smoker',hue='sex',row='time',aspect=0.6)
```
