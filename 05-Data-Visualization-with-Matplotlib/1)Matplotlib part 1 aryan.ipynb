```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL METHOD to create a MATPLOTLIB plot
```

```python
plt.plot(x,y)#same as x vs y
#plt.show() if using another ide
```

```python
plt.plot(x,x**10)
plt.xlabel('X label')
plt.ylabel('Y label ')
plt.title('OUR GRAPH')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
#SUBPLOT is creating multiple plots on the same canvas
```

```python
#OBJECT ORIENTED METHOD TO CREATE MATPLOTLIB PLOT
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.5,0.8,0.2])
#the list takes 4 arguments here left,bottom,width,height ranging from 0 to 1
axes.plot(x,y)

#add x label and y label
axes.set_xlabel('X LABEL')
axes.set_ylabel('Y LABEL')
axes.set_title('OUR GRAPH')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.15,0.45,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title('smaller plot')
axes1.set_title('bigger plot')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.2,0.9,0.4])
axes1.plot(y,y**(1/2))
```
