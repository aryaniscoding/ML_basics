```python
#PLOT APPEARANCE
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue',linewidth=1,alpha=0.5,linestyle='--',marker='+',markersize=10,
       markerfacecolor='yellow',markeredgewidth=3,markeredgecolor = 'green')
#rgb hex codes
#alpha for line visibility

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
