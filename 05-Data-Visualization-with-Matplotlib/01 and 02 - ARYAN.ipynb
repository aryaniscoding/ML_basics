```python
#object oriented method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.1,1.2])
axes.plot(x,y)

axes.set_xlabel("x label")
axes.set_ylabel("y label")
axes.set_title("titel lesgo")
```

```python
fig2 = plt.figure()
axes1 = fig2.add_axes([0.1,0.2,1,2])
axes2 = fig2.add_axes([0.2,1.1,0.4,0.5])
axes1.plot(x,y)
axes2.plot(y,x,'r')
#(left,bottom,width,height)
```

```python
fig = plt.figure(figsize=(3,1),dpi = 100)
#dpi is dots per inches
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
#using subplots
fig,axes = plt.subplots(nrows = 1,ncols = 2,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x,'r')
plt.tight_layout()
```
